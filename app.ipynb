```python
# Page config
st.set_page_config(page_title="Fake News Detector", page_icon="📰", layout="centered")

# Title
st.markdown("<h1 style='text-align: center;'>📰 Fake News Detector</h1>", unsafe_allow_html=True)
st.write("Enter a news article below to check if it is REAL or FAKE.")

# Load model
model = pickle.load(open("model/model.pkl", "rb"))
vectorizer = pickle.load(open("model/vectorizer.pkl", "rb"))

# Clean function
def clean_text(text):
    text = re.sub(r'\W', ' ', str(text))
    text = re.sub(r'\s+', ' ', text)
    return text.lower()

# Text area
user_input = st.text_area("Paste News Article Here", height=200)

if st.button("🔍 Predict"):

    if user_input.strip() == "":
        st.warning("Please enter some news text.")
    else:
        cleaned = clean_text(user_input)
        vectorized = vectorizer.transform([cleaned])
        prediction = model.predict(vectorized)[0]
        probability = model.predict_proba(vectorized)[0]

        if prediction == 1:
            st.success(f"✅ REAL News\nConfidence: {round(probability[1]*100,2)}%")
        else:
            st.error(f"🚨 FAKE News\nConfidence: {round(probability[0]*100,2)}%")
```

```python
def clean_text(text):
    text = re.sub(r'\W', ' ', str(text))
    text = re.sub(r'\s+', ' ', text)
    return text.lower()
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import re

# Text cleaning
def clean_text(text):
    text = re.sub(r'\W', ' ', str(text))
    text = re.sub(r'\s+', ' ', text)
    return text.lower()

# Create UI elements
text_area = widgets.Textarea(
    placeholder='Paste your news article here...',
    layout=widgets.Layout(width='100%', height='200px')
)

button = widgets.Button(
    description="🔍 Predict",
    button_style='primary'
)

output = widgets.Output()

# Prediction function
def on_button_click(b):
    with output:
        clear_output()
        user_input = text_area.value
        
        if user_input.strip() == "":
            print("⚠️ Please enter some text.")
            return
        
        cleaned = clean_text(user_input)
        vectorized = vectorizer.transform([cleaned])
        prediction = model.predict(vectorized)[0]
        probability = model.predict_proba(vectorized)[0]
        
        if prediction == 1:
            print("✅ REAL News")
            print("Confidence:", round(probability[1]*100,2), "%")
        else:
            print("🚨 FAKE News")
            print("Confidence:", round(probability[0]*100,2), "%")

button.on_click(on_button_click)

display(text_area, button, output)
```
